## Warehouse optimization

```python
import numpy as np
from docplex.mp.model import Model
```

```python
distance_map = {
    'Rochester': 150,
    'Buffalo': 280,
    'Syracruse': 155,
    'Albany': 200,
    'Utica': 180,
    'New Rochelle': 50,
    'Schenectady': 300,
    'Yonkers': 80,
    'Mount Vernon': 240,
    'Hempstead': 25,
    'Webster': 110
}
```

```python
warehouse_df['distance_to_store'] = (
    warehouse_df['warehouse_location']
    .map(distance_map)
    .fillna(1500)
)
```

```python
warehouse_df['transport_cost_per_unit'] = (
    warehouse_df['shipping_cost_usd']/warehouse_df['stock_quantity']
        .replace(0, 1)
)
```

```python
warehouse_df['lead_time_days'] = (warehouse_df['delivery_days'])
```

```python
warehouse_opt_df = warehouse_df[['warehouse_id','warehouse_location','product_id','product_name', 'category',
       'sub_category', 'brand','stock_quantity','distance_to_store','transport_cost_per_unit','lead_time_days'
]].rename(columns={
    'stock_quantity': 'available_stock'
})
```

```python
forecast_df = pd.read_csv('inventory_forecast_output.csv', on_bad_lines='skip')
```

```python
reorder_df = forecast_df[forecast_df['reorder_quantity'] > 0].copy()
```

```python
mdl = Model(name='inventory_optimization')
```

```python
x = {}

for _, r in reorder_df.iterrows():

    p = r['product_id']

    candidate_wh = warehouse_opt_df[
        warehouse_opt_df['product_id'] == p
    ]

    for _, w in candidate_wh.iterrows():

        key = (w['warehouse_id'], p)

        x[key] = mdl.continuous_var(
            lb=0,
            name=f"x_{w['warehouse_id']}_{p}"
        )
```

```python
objective_terms = []

for _, w in warehouse_opt_df.iterrows():

    key = (w['warehouse_id'], w['product_id'])

    # Only use variables that exist
    if key in x:

        cost = (

            w['transport_cost_per_unit']

            + 0.01 * w['distance_to_store']

            + 2 * w['lead_time_days']

        )

        objective_terms.append(

            x[key] * cost

        )

objective = mdl.sum(objective_terms)

mdl.minimize(objective)
```

```python
for _, r in reorder_df.iterrows():

    p = r['product_id']

    reorder_qty = r['reorder_quantity']

    candidate_wh = warehouse_opt_df[
        warehouse_opt_df['product_id'] == p
    ]

    mdl.add_constraint(

        mdl.sum(

            x[(w['warehouse_id'], p)]

            for _, w in candidate_wh.iterrows()

        )

        >=

        reorder_qty

    )
```

```python
for _, w in warehouse_opt_df.iterrows():

    key = (w['warehouse_id'], w['product_id'])

    if key in x:

        mdl.add_constraint(

            x[key]

            <=

            w['available_stock']

        )
```

```python
solution = mdl.solve(log_output=True)

print(solution)
```

```python
results = []

for (warehouse_id, product_id), var in x.items():

    qty = var.solution_value

    if qty is not None and qty > 0:

        # Warehouse information
        wh = warehouse_opt_df[
            (warehouse_opt_df['warehouse_id'] == warehouse_id)
            &
            (warehouse_opt_df['product_id'] == product_id)
        ].iloc[0]

        # Forecast / reorder information
        reorder_info = reorder_df[
            reorder_df['product_id'] == product_id
        ].iloc[0]

        # Product information
        product_info = warehouse_df[
            warehouse_df['product_id'] == product_id
        ].iloc[0]

        results.append({

            # Warehouse fields
            'warehouse_id': warehouse_id,

            'warehouse_location':
                wh['warehouse_location'],

            # Product fields
            'product_id': product_id,

            'product_name':
                product_info['product_name'],

            'category':
                product_info['category'],

            'sub_category':
                product_info['sub_category'],

            'brand':
                product_info['brand'],

            # Optimization output
            'shipment_quantity':
                round(qty, 2),

            'warehouse_available_stock':
                wh['available_stock'],

            'distance_to_store':
                wh['distance_to_store'],

            'lead_time_days':
                wh['lead_time_days'],

            # Forecast fields
            'predicted_demand':
                reorder_info['predicted_demand'],

            'reorder_quantity':
                reorder_info['reorder_quantity']

        })
```

```python
optimized_replenishment_df = pd.DataFrame(results)

optimized_replenishment_df = (

    optimized_replenishment_df

    .sort_values(
        by=['product_id', 'shipment_quantity'],
        ascending=[True, False]
    )

)
```

```python
optimized_replenishment_df = (
    optimized_replenishment_df
    .reset_index(drop=True)
)
optimized_replenishment_df.head(50)
```

```python
optimized_replenishment_df.to_csv(
    'optimized_replenishment_plan.csv',
    index=False
)

print("Optimization complete.")
```
